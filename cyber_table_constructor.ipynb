```python
csv_file_path = '../../data/stocknames.csv.gz'

# Read the CSV file into a pandas DataFrame
stock_info = pd.read_csv(csv_file_path)

url_columns = [col for col in stock_info.columns if 'url' in col]
stock_info.drop(columns=url_columns, inplace=True)


# Display the DataFrame
display(stock_info)
```

```python
#create template for each score
# Generate index ranging from January 2007 to December 2024
dates = pd.date_range(start='2007-01-01', end='2024-12-31', freq='ME').strftime('%Y-%m')

# Create DataFrame with NaN values
template = pd.DataFrame(index=dates, columns=ticker_list)
template[:] = np.nan
```

```python
def date_finder(ticker,year):
    column_filling_name="filling_date_"+str(year)
    # Find the corresponding value in column A where column B matches the value of reference
    date=stock_info.loc[stock_info['ticker'] == ticker, column_filling_name].values[0]

    if not isinstance(date, str):
        print(ticker)
        print(year)
        STOP

    return '-'.join(date.split('-')[:2])
```

```python
def cyber_score_template_construction(temp,option="Overall"):
    #option="Overall"      :  temp must be a COPY of template
    #option="Sentiment"    :  temp must be a COPY of template
    #option="14"           :  temp must be a dict containing 14 COPY of template
    #option="SuperTactic"  :  temp must be a a dict containing 4 COPY of template
    


    if option=="Overall":
        if isinstance(temp, pd.DataFrame):
            cyber_template=temp.copy(deep=True)

        else:
            stop
                
    elif option=="Sentiment":
        if isinstance(temp, pd.DataFrame):
            cyber_template=temp.copy(deep=True)

        else:
            stop
        
    elif option=="14":
        if isinstance(temp, dict):
            cyber_dict_template = copy.deepcopy(temp)
        else:
            stop
        
    elif option=="SuperTactic":
        if isinstance(temp, dict):
            cyber_dict_template = copy.deepcopy(temp)
        else:
            stop

    histo_dict={}
    
    years=np.arange(2007,2025) #!!!!!!!!
    for year in years:
        print("starting year ",year," ...")
        directory='../../data/10k_statements_new/similarity_with_MITRES/'+str(year)+'/'
    
    
    
        if option=="Overall":
            dict_cyber_score_for_this_year={}
                
        elif option=="Sentiment":
            dict_cyber_score_for_this_year={}
            
        elif option=="14":
            nested_dict_cyber_score_for_this_year={}
            
        elif option=="SuperTactic":
            nested_dict_cyber_score_for_this_year={}
        
    
    
        
        for matrix_name in tqdm(os.listdir(directory)):
            ticker=matrix_name.split('_')[0]
            if "TSLA" in matrix_name or "AAPL" in matrix_name or True:
                arr=np.load(directory+matrix_name)
                cosim=arr['arr']
        
               
        
                if option=="Overall":
                    
                    histo_cyber=np.max(cosim,axis=1)
                    cyber_score=select_cyber_score_from_histo(histo_cyber)
    
                    dict_cyber_score_for_this_year[ticker]=cyber_score

                    date_to_add=date_finder(ticker,year)
                    
                    cyber_template.loc[date_to_add, ticker] = cyber_score
    
                
                elif option=="Sentiment":
                    histo_cyber=np.max(cosim,axis=1)
                  
                    
                    directory_sentiment='../../data/10k_statements_new/risk_occurences/'+str(year)+'/'+ticker+'_risk_occurences.txt'
                    sentiment_array=np.loadtxt(directory_sentiment)
                    sentiment_array[sentiment_array>0]=1

                    cyber_score=np.max(histo_cyber*sentiment_array)
                    dict_cyber_score_for_this_year[ticker]=cyber_score

                    date_to_add=date_finder(ticker,year)
                    
                    cyber_template.loc[date_to_add, ticker] = cyber_score

                    
                    
                    
                elif option=="14":
                    dict_cosim=splitter_of_cos_matrix_into_tactics(cosim)

                    date_to_add=date_finder(ticker,year)

                    for tactic in list(dict_cosim.keys()):
                        cyber_score=select_cyber_score_from_histo(np.max(dict_cosim[tactic],axis=1))
                        

                        cyber_dict_template[tactic].loc[date_to_add, ticker] = cyber_score
    
                    
                    
                elif option=="SuperTactic":
                    dict_cosim=splitter_of_cos_matrix_into_tactics(cosim)
    
                    dict_cosim=organize_in_super_tactic(dict_cosim) 
    
                    date_to_add=date_finder(ticker,year)

                    for tactic in list(dict_cosim.keys()):
                        cyber_score=select_cyber_score_from_histo(np.max(dict_cosim[tactic],axis=1))
                        

                        cyber_dict_template[tactic].loc[date_to_add, ticker] = cyber_score
    
                   
                    
          
    if option=="Overall":
            return cyber_template
                
    elif option=="Sentiment":
            return cyber_template
            
    elif option=="14":
            return cyber_dict_template
            
    elif option=="SuperTactic":
            return cyber_dict_template

    
 
```

```python
csv_file_path = '../../data/stocknames.csv.gz'

# Read the CSV file into a pandas DataFrame
stock_info = pd.read_csv(csv_file_path)

url_columns = [col for col in stock_info.columns if 'url' in col]
stock_info.drop(columns=url_columns, inplace=True)


# Display the DataFrame
display(stock_info)
```

```python
ticker_list=stock_info.ticker.values
```

```python
#create template for each score
# Generate index ranging from January 2007 to December 2024
dates = pd.date_range(start='2007-01-01', end='2024-12-31', freq='ME').strftime('%Y-%m')

# Create DataFrame with NaN values
template = pd.DataFrame(index=dates, columns=ticker_list)
template[:] = np.nan
```

```python
def date_finder(ticker,year):
    column_filling_name="filling_date_"+str(year)
    # Find the corresponding value in column A where column B matches the value of reference
    date=stock_info.loc[stock_info['ticker'] == ticker, column_filling_name].values[0]

    if not isinstance(date, str):
        print(ticker)
        print(year)
        STOP

    return '-'.join(date.split('-')[:2])
```

```python
def textual_score_template_construction(temp):


    readability_table=temp.copy(deep=True)
    secret_table=temp.copy(deep=True)
    risk_length_table=temp.copy(deep=True)
    

    histo_dict={}
    
    years=np.arange(2007,2025) #!!!!!!!!
    for year in years:
        print("starting year ",year," ...")
        directory='../../data/10k_statements_new/financial_text_variables/'+str(year)+'/'

        
        for matrix_name in tqdm(os.listdir(directory)):
            ticker=matrix_name.split('_')[0]
            if "TSLA" in matrix_name or "AAPL" in matrix_name or True:
                triple_score = np.load(directory+matrix_name)


                date_to_add=date_finder(ticker,year)
                

                readability_table.loc[date_to_add, ticker] = triple_score[0]
                secret_table.loc[date_to_add, ticker] = triple_score[1]
                risk_length_table.loc[date_to_add, ticker] = triple_score[2]
                
                            
    return readability_table, secret_table, risk_length_table
    
                   
```

```python
readability_table, secret_table, risk_length_table=textual_score_template_construction(template.copy(deep=True))
```

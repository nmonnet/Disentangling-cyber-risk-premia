# Correlation Table

```python
def to_series(df_,name=None):
    df=df_.copy(deep=True)
    # Reshape the DataFrame into a Series and manipulate the index
    series = df.T.stack()
    series.index = series.index.map(lambda x: f"{x[1]}{x[0]}")

    series.name=name
    return series
```

```python
def statistical_table_for_1_cyberscore(cyberscore):
    cyberscore_series=to_series(cyberscore)

    dict_series={}
    for key in list(dict_df.keys()):
        dict_series[key]=to_series(dict_df[key])


    correlations = []
    mean_values = []
    std_values = []
    min_values = []
    max_values = []
    p1_values = []
    p25_values = []
    p50_values = []
    p75_values = []
    p99_values = []
    keys_list = []
    
    for key in dict_series.keys():
        merged = cyberscore_series.to_frame().merge(dict_series[key].to_frame(), left_index=True, right_index=True, how='inner')
        merged.dropna(inplace=True)
        
        correlation = merged['0_x'].corr(merged['0_y'])
        correlations.append(np.round(correlation, 2))
        
        mean_values.append(np.round(np.mean(dict_series[key]), 2))
        std_values.append(np.round(np.std(dict_series[key]), 2))
        min_values.append(np.round(np.min(dict_series[key]), 2))
        max_values.append(np.round(np.max(dict_series[key]), 2))
        p1_values.append(np.round(np.percentile(dict_series[key], 1), 2))
        p25_values.append(np.round(np.percentile(dict_series[key], 25), 2))
        p50_values.append(np.round(np.percentile(dict_series[key], 50), 2))
        p75_values.append(np.round(np.percentile(dict_series[key], 75), 2))
        p99_values.append(np.round(np.percentile(dict_series[key], 99), 2))
        
        keys_list.append(key)

    # Set display options
    pd.set_option('display.float_format', lambda x: '%.2f' % x)

    result_df = pd.DataFrame({
        'Mean': mean_values,
        'Std': std_values,
        'Min': min_values,
        'Max': max_values,
        'P1': p1_values,
        'P25': p25_values,
        'P50': p50_values,
        'P75': p75_values,
        'P99': p99_values,
        'Correlation': correlations
    }, index=keys_list)
    
    return result_df
```

```python
def convert_index_to_industry_FF(x):
   
    if np.isnan(x):
        return "Other"
    else:
        index=int(x-1)
        return industries_index[index]
```

```python
file_path = 'Financial_ratios/FF_industry_code.csv'
FF_table=pd.read_csv(file_path)
FF_table.rename(columns={'Unnamed: 0': 'ticker'}, inplace=True)
FF_table["FF"]=FF_table["0"].apply(lambda x : convert_index_to_industry_FF(x))
FF_table.drop("0", axis=1, inplace=True)
```

```python
def cyberscore_per_industry_mean():

    index_list=industries_index
    columns_list=[item.replace('_', ' ') for item in list(cyberscore_all_dict.keys())]
    
    mean_table=pd.DataFrame(index=index_list, columns=columns_list, dtype=float)
    
    for cyber_key in list(cyberscore_all_dict.keys()):
    
        
        for ff in industries_index:
            group_of_ticker=list(FF_table[FF_table["FF"]==ff]["ticker"])
    
        
    
            averaged_cyb=np.nanmean(cyberscore_all_dict[cyber_key][group_of_ticker].values)

           
    
            
            mean_table.loc[ff, cyber_key.replace('_', ' ')] = np.round(averaged_cyb,3)
    
    
    
    return mean_table
    


```

```python
def classify_percentage(percentage):
    if percentage is None or percentage != percentage:  # Check for NaN
        return np.nan
    elif percentage < 0.2:
        return 'P1'
    elif 0.2 <= percentage < 0.4:
        return 'P2'
    elif 0.4 <= percentage < 0.6:
        return 'P3'
    elif 0.6 <= percentage < 0.8:
        return 'P4'
    else:
        return 'P5'
```

```python
def columns_containing_Pk(df,Pk, index_label):
    row_index = df.index.get_loc(index_label)
    columns_with_Pk = [column for column in df.columns if df.loc[index_label, column] == Pk]
    return columns_with_Pk
```

```python
def returns_sub_portfolio(list_ticker,date,mkt_strat):
    tot_market_caps=0
    array_weight=[]
    return_array=[]
    for tick in list_ticker:
        array_weight.append(mkt_strat.loc[date, tick])
        tot_market_caps+=mkt_strat.loc[date, tick]

        return_array.append(returns_monthly.loc[date, tick])

    array_weight=np.array(array_weight)/tot_market_caps
    return_array=np.array(return_array)

   
    return np.sum(return_array*array_weight)
        
```

```python
def quarter_mask(ret_,cap_):
    #this function aim to produce a mask of availability of data used to construct the portfolio
    #basically it checks that market caps of end of the ending month of last quarter is available and returns is available
    #monthly for the whole quarter
    ret=ret_.copy(deep=True)
    cap=cap_.copy(deep=True).shift(periods=1)

    mask=pd.DataFrame(index=ret.index, columns=ret.columns).fillna(1)
    
    
    for ticker in mask.columns:
        mask.loc[mask.index[0], ticker]=np.nan
        index=1
        while index+2<mask.shape[0]:
            
            if np.isnan(ret[ticker][index]*ret[ticker][index+1]*ret[ticker][index+2]):
                mask.loc[mask.index[index], ticker]=np.nan
                mask.loc[mask.index[index+1], ticker]=np.nan
                mask.loc[mask.index[index+2], ticker]=np.nan
                
            if np.isnan(cap[ticker][index]):
                mask.loc[mask.index[index], ticker]=np.nan
                mask.loc[mask.index[index+1], ticker]=np.nan
                mask.loc[mask.index[index+2], ticker]=np.nan
                
                
            index+=3

        #two last index
        if np.isnan(ret[ticker][index]*ret[ticker][index+1]):
                mask.loc[mask.index[index], ticker]=np.nan
                mask.loc[mask.index[index+1], ticker]=np.nan
                
                
        if np.isnan(cap[ticker][index]):
                mask.loc[mask.index[index], ticker]=np.nan
                mask.loc[mask.index[index+1], ticker]=np.nan
              
    # Count non-NaN values in each column
    non_nan_counts = mask.count()
    
    # Identify columns with less than 36 non-NaN values
    columns_to_fill = non_nan_counts[non_nan_counts < 36].index
    
    # Fill identified columns with NaN
    mask[columns_to_fill] = np.nan




    #eliminate small caps firms


    # Check maximum values of each column in cap
    max_values_cap = cap.max()
    
    # Iterate over columns of df1
    for column in mask.columns:
        # Check if the max value in cap is below xxx millions
        if max_values_cap[column] < -300000000:
            # Replace values in mask column with NaN
            mask[column] = np.nan
        
    return mask
```

```python
def cyber_portfolio(cyberscore_):
    #construct the mask of availability
    mask_available=quarter_mask(returns_monthly, market_caps)
    
    #
    cyberscore=cyberscore_.copy(deep=True)
    #check that the mask and the cyberscore table ranged from 2007-01 to 2023-12
    cyberscore=cyberscore[~cyberscore.index.str.contains('2024')]


    #complete the cyberscore table with last known values (since they once per year)
    cyberscore=backward_completion_cyberscore(cyberscore)
    cyberscore_shifted=cyberscore.shift(periods=1)

    #quarterlization 
    # Step 1: Create a boolean mask based on the index conditions
    index_conditions = [ind.endswith(('02', '05', '08', '11')) for ind in cyberscore_shifted.index]
        
    # Step 2: Use the boolean mask to select rows
    rows_to_set_nan = cyberscore_shifted[np.logical_not(index_conditions)]
        
    # Step 3: Set the selected rows to NaN
    cyberscore_shifted.loc[rows_to_set_nan.index] = np.nan

    # Step 4: fill the nan void by repeating the value of cyberscore at the end of the month of the end of last quarter ('02', '05', '08', '11')
    cyberscore_strategy=cyberscore_shifted.ffill()

    #remove unavailable values due to missing returns or market caps
    cyberscore_strategy=cyberscore_strategy.mul(mask_available) #(either multiply by 1 or Nan)

    
    cyberfolio_class=cyberscore_strategy.rank(pct=True, axis=1, method="first")
    cyberfolio_class=cyberfolio_class.map(classify_percentage)
    

    

    #define the market caps that will be used to define the weight inside each portfolio
    #----------------------------------------------------------------------------------------

    market_caps_shifted = market_caps.shift(periods=1)

    
    # Step 1: Create a boolean mask based on the index conditions
    index_conditions = [ind.endswith(('02', '05', '08', '11')) for ind in market_caps_shifted.index]
    
    # Step 2: Use the boolean mask to select rows
    rows_to_set_nan = market_caps_shifted[np.logical_not(index_conditions)]
    
    # Step 3: Set the selected rows to NaN
    market_caps_shifted.loc[rows_to_set_nan.index] = np.nan
    
    market_caps_strategy = market_caps_shifted.ffill()
   
    

    #define which firms will be part of the returns of each portfolio
    #-------------------------------------------------------------------------------------------
    class_p_name=["P1","P2","P3","P4","P5"]
  

    p_df=pd.DataFrame(index=cyberfolio_class.index, columns=class_p_name)
    
    for date in cyberfolio_class.index:
        for class_p in class_p_name:
            list_ticker_in_class_p=columns_containing_Pk(cyberfolio_class,class_p,date)
            
            p_df.loc[date, class_p]=list_ticker_in_class_p


    
    p_df.replace(to_replace=[], value=np.nan, inplace=True)
    p_df.dropna(inplace=True)
    p_df = p_df.drop(p_df.index[0])

    output_portfolio_returns=pd.DataFrame(index=p_df.index, columns=p_df.columns)


    for class_p in class_p_name:
        for date in output_portfolio_returns.index:
            output_portfolio_returns.loc[date, class_p]=returns_sub_portfolio(p_df.loc[date, class_p],date,market_caps_strategy)
        
            
    
    

    return output_portfolio_returns, p_df
```

```python
def t_test(mean_array, std_array,n):
    dof=n-1
    # Calculate t-statistics
    t_values = mean_array / (std_array / np.sqrt(n))

    # Calculate p-values
    p_values = 2 * (1 - t.cdf(np.abs(t_values), dof))

    return p_values
```

```python
def classify_percentage_20(percentage):
    if percentage is None or percentage != percentage:  # Check for NaN
        return np.nan
    elif percentage < 0.05:
        return 'P1'
    elif 0.05 <= percentage < 0.1:
        return 'P2'
    elif 0.1 <= percentage < 0.15:
        return 'P3'
    elif 0.15 <= percentage < 0.2:
        return 'P4'
    elif 0.2 <= percentage < 0.25:
        return 'P5'
    elif 0.25 <= percentage < 0.3:
        return 'P6'
    elif 0.3 <= percentage < 0.35:
        return 'P7'
    elif 0.35 <= percentage < 0.4:
        return 'P8'
    elif 0.4 <= percentage < 0.45:
        return 'P9'
    elif 0.45 <= percentage < 0.5:
        return 'P10'
    elif 0.5 <= percentage < 0.55:
        return 'P11'
    elif 0.55 <= percentage < 0.6:
        return 'P12'
    elif 0.6 <= percentage < 0.65:
        return 'P13'
    elif 0.65 <= percentage < 0.7:
        return 'P14'
    elif 0.7 <= percentage < 0.75:
        return 'P15'
    elif 0.75 <= percentage < 0.8:
        return 'P16'
    elif 0.8 <= percentage < 0.85:
        return 'P17'
    elif 0.85 <= percentage < 0.9:
        return 'P18'
    elif 0.9 <= percentage < 0.95:
        return 'P19'
    else:
        return 'P20'
```

```python
def cyber_sub_portfolio(list_ticker,date,cyb_score,mkt_strat):
    tot_market_caps=0
    array_weight=[]
    cyb_array=[]
    for tick in list_ticker:
        array_weight.append(mkt_strat.loc[date, tick])
        tot_market_caps+=mkt_strat.loc[date, tick]

        cyb_array.append(cyb_score.loc[date, tick])

    array_weight=np.array(array_weight)/tot_market_caps
    cyb_array=np.array(cyb_array)

   
    return np.sum(cyb_array*array_weight)
```

```python
def cyber_portfolio_20(cyberscore_):
    #construct the mask of availability
    mask_available=quarter_mask(returns_monthly, market_caps)
    
    #
    cyberscore=cyberscore_.copy(deep=True)
    #check that the mask and the cyberscore table ranged from 2007-01 to 2023-12
    cyberscore=cyberscore[~cyberscore.index.str.contains('2024')]


    #complete the cyberscore table with last known values (since they once per year)
    cyberscore=backward_completion_cyberscore(cyberscore)
    cyberscore_shifted=cyberscore.shift(periods=1)

    #quarterlization 
    # Step 1: Create a boolean mask based on the index conditions
    index_conditions = [ind.endswith(('02', '05', '08', '11')) for ind in cyberscore_shifted.index]
        
    # Step 2: Use the boolean mask to select rows
    rows_to_set_nan = cyberscore_shifted[np.logical_not(index_conditions)]
        
    # Step 3: Set the selected rows to NaN
    cyberscore_shifted.loc[rows_to_set_nan.index] = np.nan

    # Step 4: fill the nan void by repeating the value of cyberscore at the end of the month of the end of last quarter ('02', '05', '08', '11')
    cyberscore_strategy=cyberscore_shifted.ffill()

    #remove unavailable values due to missing returns or market caps
    cyberscore_strategy=cyberscore_strategy.mul(mask_available) #(either multiply by 1 or Nan)

    
    cyberfolio_class=cyberscore_strategy.rank(pct=True, axis=1, method="first")
    cyberfolio_class=cyberfolio_class.map(classify_percentage_20)
    

    

    #define the market caps that will be used to define the weight inside each portfolio
    #----------------------------------------------------------------------------------------

    market_caps_shifted = market_caps.shift(periods=1)

    
    # Step 1: Create a boolean mask based on the index conditions
    index_conditions = [ind.endswith(('02', '05', '08', '11')) for ind in market_caps_shifted.index]
    
    # Step 2: Use the boolean mask to select rows
    rows_to_set_nan = market_caps_shifted[np.logical_not(index_conditions)]
    
    # Step 3: Set the selected rows to NaN
    market_caps_shifted.loc[rows_to_set_nan.index] = np.nan
    
    market_caps_strategy = market_caps_shifted.ffill()
   
    

    #define which firms will be part of the returns of each portfolio
    #-------------------------------------------------------------------------------------------
    class_p_name = ["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10",
                "P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]


    p_df=pd.DataFrame(index=cyberfolio_class.index, columns=class_p_name)
 
    
    for date in cyberfolio_class.index:
        for class_p in class_p_name:
            list_ticker_in_class_p=columns_containing_Pk(cyberfolio_class,class_p,date)
            
            p_df.loc[date, class_p]=list_ticker_in_class_p


    
    p_df.replace(to_replace=[], value=np.nan, inplace=True)
    p_df.dropna(inplace=True)
    p_df = p_df.drop(p_df.index[0])

    output_portfolio_returns=pd.DataFrame(index=p_df.index, columns=p_df.columns)

    output_portfolio_cyber=pd.DataFrame(index=p_df.index, columns=p_df.columns)

    cyb_to_use=backward_completion_cyberscore(cyberscore_.copy(deep=True))

    for class_p in class_p_name:
        for date in output_portfolio_returns.index:
            output_portfolio_returns.loc[date, class_p]=returns_sub_portfolio(p_df.loc[date, class_p],date,market_caps_strategy)

            output_portfolio_cyber.loc[date, class_p]=cyber_sub_portfolio(p_df.loc[date, class_p],date,cyb_to_use,market_caps_strategy)
        
            
    
    

    return output_portfolio_returns, p_df, output_portfolio_cyber, market_caps_strategy
```

```python
def build_beta_tables(model_name):


    if os.path.exists("beta_tables/"+model_name):
        # Get the list of files in the folder
        files = os.listdir("beta_tables/"+model_name)

        dict_factor_beta={}
        # Iterate over each file
        for file in files:
            # Check if the file is a CSV file
            if file.endswith('.csv'):
                # Extract the name of the CSV file (without the extension)
                name_of_csv = os.path.splitext(file)[0]
                
                # Construct the full file path
                file_path = os.path.join("beta_tables/"+model_name, file)
                
                # Read the CSV file into a DataFrame
                df = pd.read_csv(file_path)
                df.set_index('date', inplace=True)
                
                # Store the DataFrame in the dictionary
                name_of_csv=name_of_csv.split('_beta')[0].replace("_","-")
                dict_factor_beta[name_of_csv] = df


    else:
        #get a table with all relevant factors 
        industry_12_portfolio=get_industry_12_portfolio()
        FF_5_Momentum=get_FF_5_Momentum()
        FF_5_factors=get_FF_5_factors()
    
        all_factors=pd.merge(industry_12_portfolio, FF_5_Momentum, left_index=True, right_index=True)
        all_factors=pd.merge(all_factors, FF_5_factors, left_index=True, right_index=True)
        
        all_factors=all_factors.drop('RF', axis=1)
    
        list_factors=list(all_factors.columns)
        #print(list_factors)
    
    
    
    
        # Ensure that both the DataFrame and the Series have the same index (dates)
        df_returns=returns_monthly.copy(deep=True)
        
        #note model_2 has not first step in Fama-McBeth
        
        if model_name=="model_1" or model_name=="model_3":
            factors_name=['Mkt-RF']

        elif model_name=="model_2":
            return None
            
        elif model_name=="model_4":
            factors_name=['UMD','Mkt-RF','SMB','HML']

        elif model_name=="model_5":
            factors_name=['Mkt-RF','SMB','HML','RMW','CMA']
            


            
        factor=all_factors[factors_name][~all_factors[factors_name].index.str.contains('2024')][522:] #only keep 2007-2023 data

    
       
        
        # Initialize an empty DataFrame to store beta coefficients
        beta_df = pd.DataFrame(index=df_returns.index, columns=df_returns.columns)

        # Initialize a dict for the empty DataFrame to store beta coefficients
        dict_factor_beta={}
        for f_n in factors_name:
            dict_factor_beta[f_n]=beta_df.copy(deep=True)
            
        
        
        # Iterate over each date in the index
        for i, date in tqdm(enumerate(df_returns.index), total=len(df_returns.index)):
            if i>=23:
                # Get the 24 most recent observations before the current date
                start_date = df_returns.index[df_returns.index <= date][-24]
                end_date = date
                
                # Filter the DataFrame to get the subset of returns for the past 24 months
                subset_returns = df_returns.loc[start_date:end_date]
                
                # Filter the factor series to get the corresponding subset
                subset_factor = factor.loc[start_date:end_date]
        
                #print(subset_returns)
                #print(subset_factor)
        
                
           
                # Add a constant to the factor for the intercept term
                X = sm.add_constant(subset_factor)
                
                # Iterate over each firm's returns
                for ticker in df_returns.columns:
                  
                    # Extract the returns for the current ticker
                    y = subset_returns[ticker]
                    
                    # Perform linear regression
                    model = sm.OLS(y, X)
                    results = model.fit()

                    #print(results.summary())
                    #print(float(results.params["Mkt-RF"]))
                    
                    # Store the beta coefficient in the beta_df
                    for f_n in factors_name:
                        dict_factor_beta[f_n].loc[date,ticker]=float(results.params[f_n])
                      
                   
                    
                    

        # Save the DataFrame to a CSV file in the "beta_tables" folder
        # Create a directory named "model_1" if it doesn't exist
        folder_name = "beta_tables/"+model_name
        if not os.path.exists(folder_name):
            os.makedirs(folder_name)
        
        # Define the file path where you want to save the DataFrame
        for f_n in factors_name:
            file_path = os.path.join(folder_name, f_n.replace("-","_")+"_beta_df.csv")
            dict_factor_beta[f_n].to_csv(file_path)




    return dict_factor_beta








        
       
```

```python
def FamaMcBeth(model_name, output_portfolio_returns_20_,output_portfolio_cyber_20_,portfolio_compo_,mkt_caps_strat_):



    portfolio_composition_20=portfolio_compo_.copy(deep=True)
    market_caps_weight_20=mkt_caps_strat_.copy(deep=True)

    returns=output_portfolio_returns_20_.copy(deep=True)
    cybers=output_portfolio_cyber_20_.copy(deep=True)

    
    if model_name=="model_1" or model_name=="model_3":
        factors_name=['Mkt-RF']

    elif model_name=="model_2":
        factors_name=[]
        
    elif model_name=="model_4":
        factors_name=['UMD','Mkt-RF','SMB','HML']

    elif model_name=="model_5":
        factors_name=['Mkt-RF','SMB','HML','RMW','CMA']
        


    
    #first, build beta_p from pre-constructed rolling beta on individual stocks
    dict_beta_i=build_beta_tables(model_name)

    
    dict_beta_p={}
    for f_n in factors_name:
        dict_beta_p[f_n]=build_beta_p(portfolio_composition_20,market_caps_weight_20,dict_beta_i,f_n)


    
    results_df=FamaMcBeth_last_step(model_name, dict_beta_p, returns,cybers)

    return results_df
```

```python
def build_beta_p(portfolio_compo_,mkt_caps_strat_,dict_beta_df,beta_name):
    
    portfolio_composition_20=portfolio_compo_.copy(deep=True)
    market_caps_weight_20=mkt_caps_strat_.copy(deep=True)
    


    beta_portfolio=pd.DataFrame(index=portfolio_composition_20.index, columns=portfolio_composition_20.columns)
    
    class_p_name = ["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10","P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]
    
    for date in portfolio_composition_20.index:
        for class_p in class_p_name:
    
            list_ticker=portfolio_composition_20.loc[date, class_p]
    
            weight_array=[]
            beta_array=[]
            
            for tick in list_ticker:
                w=market_caps_weight_20.loc[date,tick]
                b=dict_beta_df[beta_name].loc[date,tick]
                if not(np.isnan(w*b)):
                    weight_array.append(w)
                    beta_array.append(b)
            if len(weight_array)==0:
                beta_p=np.nan
    
            else:
                weight_array=np.array(weight_array)
                beta_array=np.array(beta_array)
                beta_p=np.sum(weight_array*beta_array/np.sum(weight_array))
                
            beta_portfolio.loc[date, class_p]=beta_p
    return beta_portfolio
    
```

```python
def FamaMcBeth_last_step(option, dict_beta_p, output_portfolio_returns_20_,output_portfolio_cyber_20_):

    returns=output_portfolio_returns_20_.copy(deep=True)
    cybers=output_portfolio_cyber_20_.copy(deep=True)

    #all table must be from 2009 to 2024
    returns=returns[~returns.index.str.contains('2007')]
    returns=returns[~returns.index.str.contains('2008')]
    returns=returns[~returns.index.str.contains('2023')]

    cybers=cybers.shift(1)
    cybers=cybers[~cybers.index.str.contains('2007')]
    cybers=cybers[~cybers.index.str.contains('2008')]
    cybers=cybers[~cybers.index.str.contains('2023')]

    dict_factor={}
    for key in dict_beta_p.keys():
        temp_dic=dict_beta_p[key].copy(deep=True)
        temp_dic=temp_dic.shift(1)
        temp_dic=temp_dic[~temp_dic.index.str.contains('2007')]
        temp_dic=temp_dic[~temp_dic.index.str.contains('2008')]
        temp_dic=temp_dic[~temp_dic.index.str.contains('2023')]

        #dict_factor[key]=temp_dic.apply(standardize_row, axis=1)
        
        dict_factor[key]=temp_dic

  
    
    if option=="model_1":
        Y_column=["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10","P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]
        X_column=['Mkt-RF']
        cyber_included=False
       
    elif option=="model_2":
        Y_column=["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10","P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]
        X_column=[]
        cyber_included=True
      
    elif option=="model_3":
        Y_column=["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10","P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]
        X_column=['Mkt-RF']
        cyber_included=True
        
    elif option=="model_4":
        Y_column=["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10","P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]
        X_column=['UMD','Mkt-RF','SMB','HML']
        cyber_included=True
        
    elif option=="model_5":
        Y_column=["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10","P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]
        X_column=['Mkt-RF','SMB','HML','RMW','CMA']
        cyber_included=True
      
    elif option=="model_6 NOT AVAILABLE":
        Y_column=["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10","P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]
        X_column=['SMB','HML','RMW','CMA','NoDur','Durbl','Manuf','Enrgy','Chems','BusEq','Telcm','Utils','Shops','Hlth ','Money','Other']
        cyber_included=True

    elif option=="model_7 NOT AVAILABLE":
        Y_column=["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10","P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20",'NoDur','Durbl','Manuf','Enrgy','Chems','BusEq','Telcm','Utils','Shops','Hlth ','Money','Other']
        X_column=['Mkt-RF','SMB','HML','RMW','CMA']
        cyber_included=True
        stop


    if cyber_included:
        dict_factor["cyber"]=cybers

    #create dataframe
    if cyber_included:
        
        time_series_df=pd.DataFrame(index=returns.index, columns=X_column+["cyber"]+["intercept"]+["R2"]+["MAPE"], dtype=float)
        
    else:

        time_series_df=pd.DataFrame(index=returns.index, columns=X_column+["intercept"]+["R2"]+["MAPE"], dtype=float)

   

    for date in time_series_df.index:
        
        # Combine X arrays into a single 2D array
        if cyber_included:
            X_arrays = [dict_factor[key].loc[date].to_numpy() for key in X_column+["cyber"]]

        else:
            X_arrays = [dict_factor[key].loc[date].to_numpy() for key in X_column]

        
        
        X = np.column_stack(X_arrays).astype(float)

      
        # Add a constant if you want to fit an intercept
        X = sm.add_constant(X)

       
        Y=returns.loc[date].to_numpy().astype(float)

        
        # Fit the model
        model = sm.OLS(Y, X).fit()


       
        r_squared_adj = model.rsquared_adj
        
   
        
        # Print the summary
        #print(model.summary())

        coefficients = np.roll(model.params, -1)
        coefficients = np.append(coefficients, r_squared_adj)
        coefficients = np.append(coefficients,np.abs(model.resid).mean())
        
 
        time_series_df.loc[date]=coefficients
        
        
    #print(time_series_df)
    # Calculate the average for each column
    avg = time_series_df.mean()
    
    # Initialize lists to store t-statistics and p-values
    t_stats = []
    p_values = []
    
    # Loop through each column to calculate t-statistic and p-value
    for column in time_series_df.columns:
        # Perform one-sample t-test against 0
        t_stat, p_value = ttest_1samp(time_series_df[column], 0)
        t_stats.append(t_stat)
        p_values.append(p_value)
    
    # Create a new dataframe to store results
    results_df = pd.DataFrame({'Average': avg, 'T-Statistic': t_stats, 'P-Value': p_values})

    #you can ignore t-stat and p-value of R2 and MAPE, this is just silly nonsense
    return results_df
    
```

```python
def portfolio_sort_for_GRS(df_sort_,cyberscore_):
    #construct the mask of availability
    mask_available=quarter_mask(returns_monthly, market_caps)

    #print("done!")
    
    
    #--------------------------------------------------------------------
    df_sort=df_sort_.copy(deep=True)
    df_sort=df_sort[~df_sort.index.str.contains('2024')]
    
    df_sort_shifted=df_sort.shift(periods=1)

    
    #quarterlization 
    # Step 1: Create a boolean mask based on the index conditions
    index_conditions = [ind.endswith(('02', '05', '08', '11')) for ind in df_sort_shifted.index]
        
    # Step 2: Use the boolean mask to select rows
    rows_to_set_nan = df_sort_shifted[np.logical_not(index_conditions)]
        
    # Step 3: Set the selected rows to NaN
    df_sort_shifted.loc[rows_to_set_nan.index] = np.nan

    # Step 4: fill the nan void by repeating the value of df_sort at the end of the month of the end of last quarter ('02', '05', '08', '11')
    df_sort_strategy=df_sort_shifted.ffill()

    #remove unavailable values due to missing returns or market caps
    df_sort_strategy=df_sort_strategy.mul(mask_available) #(either multiply by 1 or Nan)


    df_sort_strategy=mask_df_sort_20(df_sort_strategy)

    

    

    #define the market caps that will be used to define the weight inside each portfolio
    #----------------------------------------------------------------------------------------

    market_caps_shifted = market_caps.shift(periods=1)

    
    # Step 1: Create a boolean mask based on the index conditions
    index_conditions = [ind.endswith(('02', '05', '08', '11')) for ind in market_caps_shifted.index]
    
    # Step 2: Use the boolean mask to select rows
    rows_to_set_nan = market_caps_shifted[np.logical_not(index_conditions)]
    
    # Step 3: Set the selected rows to NaN
    market_caps_shifted.loc[rows_to_set_nan.index] = np.nan
    
    market_caps_strategy = market_caps_shifted.ffill()
   

  
    #define which firms will be part of the returns of each portfolio
    #-------------------------------------------------------------------------------------------
    class_s_name=np.array(["S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9", "S10",
                           "S11", "S12", "S13", "S14", "S15", "S16", "S17", "S18", "S19", "S20"])
   

    p_df=pd.DataFrame(index=df_sort_strategy.index, columns=class_s_name)
 
    
    for date in df_sort_strategy.index:
        for class_s in class_s_name:
            list_ticker_in_class_s=columns_containing_Pk(df_sort_strategy,class_s,date)
            
            p_df.loc[date, class_s]=list_ticker_in_class_s


    
    p_df.replace(to_replace=[], value=np.nan, inplace=True)
    p_df.dropna(inplace=True)
    p_df = p_df.drop(p_df.index[0])

    output_portfolio_returns=pd.DataFrame(index=p_df.index, columns=p_df.columns)

    output_portfolio_cyber=pd.DataFrame(index=p_df.index, columns=p_df.columns)

    cyb_to_use=backward_completion_cyberscore(cyberscore_.copy(deep=True))
    

    for class_s in class_s_name:
        for date in output_portfolio_returns.index:
            output_portfolio_returns.loc[date, class_s]=returns_sub_portfolio(p_df.loc[date, class_s],date,market_caps_strategy)

            output_portfolio_cyber.loc[date, class_s]=cyber_sub_portfolio(p_df.loc[date, class_s],date,cyb_to_use,market_caps_strategy)
            
    
    

    return output_portfolio_returns, p_df, output_portfolio_cyber
```

```python
def GRS_for_pipeline(returns_p_cyber, returns_p_mkt, returns_p_size, returns_p_book,cyber_factor_portfolio_P1_P5,include_2023=False):
    #first get the FF5 
    start=23
    end=-12
    if include_2023:
        end=None
    
    FF_5_factors=get_FF_5_factors()
    #FF_5_Momentum=get_FF_5_Momentum()

    all_factors=pd.merge(cyber_factor_portfolio_P1_P5, FF_5_factors, left_index=True, right_index=True).astype(float)
    #all_factors=pd.merge(all_factors, FF_5_Momentum, left_index=True, right_index=True).astype(float)
    RF=all_factors["RF"].copy(deep=True).astype(float)[start:end]
    all_factors["cyber"]=all_factors["P5"]-all_factors["P1"]
    all_factors.drop(columns=["P1","P2","P3","P4","P5","RF"], inplace=True)
    all_factors_no_cyber=all_factors.drop(columns=["cyber"])

    all_factors=all_factors[start:end]
    all_factors_no_cyber=all_factors_no_cyber[start:end]
    
    column_mapping = {'S'+str(i): 'P'+str(i) for i in range(1, 21)}
    
    
    dict_returns={}
    dict_returns["cyber"]=returns_p_cyber.astype(float).rename(columns=column_mapping)[start:end].sub(RF.values, axis=0)
    dict_returns["mkt"]=returns_p_mkt.astype(float).rename(columns=column_mapping)[start:end].sub(RF.values, axis=0)
    dict_returns["size"]=returns_p_size.astype(float).rename(columns=column_mapping)[start:end].sub(RF.values, axis=0)
    dict_returns["book"]=returns_p_book.astype(float).rename(columns=column_mapping)[start:end].sub(RF.values, axis=0)

  
    
    dict_results={}
    dict_results_no_cyber={}


    class_p_name = ["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10",
                "P11", "P12", "P13", "P14", "P15", "P16", "P17", "P18", "P19", "P20"]
    
    for sorting in dict_returns.keys():
        alpha_list=[]
        residual_list=[]
        r2_list=[]
        for class_p in class_p_name:

            #with cyberfactors
           
            df_x = sm.add_constant(all_factors)

            # Extract y values
            
            y = dict_returns[sorting][class_p]
            
            # Create and fit the OLS model
            model = sm.OLS(y, df_x)
            results = model.fit()

            # Get the intercept
            alpha_list.append(results.params['const'])
            
            # Get the residuals
            residual_list.append(list(results.resid))

            r2_list.append(results.rsquared_adj)

        dict_results[sorting]=[np.array(alpha_list),np.array(residual_list),np.array(r2_list)]


        alpha_list=[]
        residual_list=[]
        for class_p in class_p_name:

            #with cyberfactors
           
            df_x = sm.add_constant(all_factors_no_cyber)
            
            # Extract y values
            y = dict_returns[sorting][class_p]
            
            # Create and fit the OLS model
            model = sm.OLS(y, df_x)
            results = model.fit()

            # Get the intercept
            alpha_list.append(results.params['const'])
            
            # Get the residuals
            residual_list.append(list(results.resid))

            r2_list.append(results.rsquared_adj)

        dict_results_no_cyber[sorting]=[np.array(alpha_list),np.array(residual_list),np.array(r2_list)]


    #with cyberfactor
    df_results=pd.DataFrame(index=list(dict_returns.keys()), columns=["GRS","p-value","R2"])
    
    for sorting in dict_returns.keys():
        T=dict_returns["cyber"].shape[0]
        N=20
        K=all_factors.shape[1]

        res=dict_results[sorting][1]
        cov_matrix = np.cov(res)
       
        inverse_cov_matrix=np.linalg.pinv(cov_matrix)
        
        alpha=dict_results[sorting][0]

     

        mean_factor=all_factors.mean().to_numpy()
        cov_factor=all_factors.cov().to_numpy()

        

        inverse_cov_factor=np.linalg.pinv(cov_factor)

     
        GRS=((T-N-K)/N)*np.dot(np.dot(alpha.T, inverse_cov_matrix), alpha)/(1+np.dot(np.dot(mean_factor.T, inverse_cov_factor), mean_factor))
        #GRS=(T/N)*((T-N-K)/(T-K-1))*np.dot(np.dot(alpha.T, inverse_cov_matrix), alpha)/(1+np.dot(np.dot(mean_factor.T, inverse_cov_factor), mean_factor))
        
        
        p_value = 1 - f.cdf(GRS, N, T-N-K)

        R2=np.mean(dict_results[sorting][2])

        df_results.loc[sorting,"GRS"]=GRS
        df_results.loc[sorting,"p-value"]=p_value
        df_results.loc[sorting,"R2"]=R2
    

    #without cyber factor
    df_results_no_cyber=pd.DataFrame(index=list(dict_returns.keys()), columns=["GRS","p-value","R2"])
    
    for sorting in dict_returns.keys():
        T=dict_returns["cyber"].shape[0]
        N=20
        K=all_factors_no_cyber.shape[1]

        res=dict_results_no_cyber[sorting][1]
        cov_matrix = np.cov(res)
       
        inverse_cov_matrix=np.linalg.pinv(cov_matrix)
        
        alpha=dict_results_no_cyber[sorting][0]


        mean_factor=all_factors_no_cyber.mean().to_numpy()
        cov_factor=all_factors_no_cyber.cov().to_numpy()

        inverse_cov_factor=np.linalg.pinv(cov_factor)

     
        GRS=(T-N-K)/N*np.dot(np.dot(alpha.T, inverse_cov_matrix), alpha)/(1+np.dot(np.dot(mean_factor.T, inverse_cov_factor), mean_factor))
        
        p_value = 1 - f.cdf(GRS, N, T-N-K)

        R2=np.mean(dict_results_no_cyber[sorting][2])

        df_results_no_cyber.loc[sorting,"GRS"]=GRS
        df_results_no_cyber.loc[sorting,"p-value"]=p_value
        df_results_no_cyber.loc[sorting,"R2"]=R2

    print("----------------------------------------------")
    print("no cyber factor included")
    print(df_results_no_cyber)
    print("")
    print("cyber factor included")
    print(df_results)
    print("")
    print("differences with-without")
    print(df_results-df_results_no_cyber)
    print("----------------------------------------------")
            
            
        
   
```

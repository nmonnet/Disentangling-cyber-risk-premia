```python
#this part test the distribution of error between the data saved by D.C. and the output of the doc2vec model saved by D.C. 
number_of_random_firms_at_each_year=3
random.seed(42)
mitres_ref_matrix=obtain_array_MITRES('../../data/MITRE_ATT&CK/descriptions/tactics_')

liste_all_value=[]
for year in np.arange(2008,2024):
    path_10k='../../data/10k_statements/tokens/'+str(year)
    path_cos_matrix_ref='../../data/cyber_risk_measures/doc2vec/per_paragraph/general/'+str(year)

    
    files = glob.glob(path_10k + '/*')
    list_available_firms_in_10k = [file.split('\\')[-1].split('_')[0] for file in files]

    files = glob.glob(path_cos_matrix_ref + '/*')
    list_available_firms_in_ref = [file.split('\\')[-1].split('_')[0] for file in files]

    list_available_firms=list(set(list_available_firms_in_10k+list_available_firms_in_ref))


    liste_min_diff=[]
    liste_max_diff=[]
    

    
    for firm in random.sample(list_available_firms, 3):
        #print(firm)
        path_firm_10k=path_10k+"/"+firm+"_tokens.csv.gz"
        path_firm_ref=path_cos_matrix_ref+"/"+firm+"_sim_matrix.csv.gz"
        try:
            array_10k=obtain_array_10k(path_firm_10k)
        except TypeError:
            print(year)
            print(firm)
            df10k=pd.read_csv(path_firm_10k)
            tokens_list = df10k.apply(lambda row: row.dropna().tolist()[1:], axis=1).tolist()

            continue
            

        cos_matrix_ref=pd.read_csv(path_firm_ref).values
        #print(cos_matrix_ref)
        
        cos_matrix_computed=compute_cos_matrix(array_10k,mitres_ref_matrix)
        cos_matrix_computed[cos_matrix_computed<0]=0 #because the matrix of references were saved that way

        diff_cos=cos_matrix_ref-cos_matrix_computed
        max_diff_score=np.max(diff_cos)
        min_diff_score=np.min(diff_cos)

        liste_min_diff.append(min_diff_score)
        liste_max_diff.append(max_diff_score)

        liste_all_value=liste_all_value+diff_cos.flatten().tolist()

    plt.hist(liste_min_diff, bins=100)  # You can adjust the number of bins as needed
    plt.xlabel("min, "+str(year)+", random draw")
    plt.ylabel('Frequency')
    #plt.yscale('log')
    plt.show()

    plt.hist(liste_max_diff, bins=100)  # You can adjust the number of bins as needed
    plt.xlabel("max, "+str(year)+", random draw")
    plt.ylabel('Frequency')
    #plt.yscale('log')
    plt.show()
    
       
plt.hist(liste_all_value, bins=100)  # You can adjust the number of bins as needed
plt.xlabel("all")
plt.ylabel('Frequency')
plt.yscale('log')
plt.show()
```

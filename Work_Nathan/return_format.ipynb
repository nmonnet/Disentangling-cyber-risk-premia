```python
series_list=[]
for ticker in tqdm(os.listdir('Financial_values/')):
    try:
        directory='Financial_values/'+ticker+'/monthly_returns.csv'
        price_table = pd.read_csv(directory)
        
        #price_table['date'] = price_table['date'].apply(lambda x: x[:-3])
        price_table.set_index('date', inplace=True)
    
        price_table[ticker]=price_table['ret']
    
        prc_table=price_table[[ticker]]


    except FileNotFoundError:

        prc_table=pd.Series(name=ticker, index=pd.Index([], name='date'))
    
    series_list.append(prc_table)
    
```

```python
return_table=pd.concat(series_list, axis=1)
return_table.sort_index(inplace=True)
```

```python
display(return_table)
save_path_ratio='Financial_ratios/'
return_table.to_csv(save_path_ratio + 'returns_monthly.csv.gz')
```
